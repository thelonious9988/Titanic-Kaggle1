```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import svm
```

```python
ds = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
id = test['PassengerId']
```

```python
ds = ds.drop(['Name','Ticket','Cabin','PassengerId'],axis = 1)
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Fit and transform the categorical column
ds['Sex'] = label_encoder.fit_transform(ds['Sex'])

print(ds)
```

```python
# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Fit and transform the categorical column
ds['Embarked'] = label_encoder.fit_transform(ds['Embarked'])

print(ds)
```

```python
def clean(ds):
    cols = ['SibSp', 'Parch', 'Fare', 'Age']
    for col in cols:
        ds[col].fillna(ds[col].median())
    ds.Embarked.fillna('U')
    return ds
ds= clean(ds)
```

```python
x = ds.drop(columns = 'Survived', axis = 1)
y = ds['Survived']
```

```python
scaler = StandardScaler()
x = scaler.fit_transform(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, stratify = y, random_state = 1)
```

```python
model = svm.SVC(kernel = 'linear')
model.fit(x_train, y_train)
```

```python
train_pred = model.predict(x_train)
train_acc = accuracy_score(train_pred, y_train)
print('Accuracy is : ', train_acc)
```

```python
test_pred = model.predict(x_test)
test_acc = accuracy_score(test_pred, y_test)
print('Accuracy is : ', test_acc)
```
